```python
#Resize images
new_imgs = []
CODZ_maps = os.listdir("CODZ_Maps")
dims = (1536,864)
for mp in CODZ_maps:
    mp = "C:\\Users\\jakek\\Desktop\\Map_Roulette\\CODZ_Maps\\" + mp
    og_img = cv2.imread(mp)
    print(f"file {mp} has been loaded via cv2 library")
    if mp != None:
        resized_img = cv2.resize(og_img, dims)
        new_imgs.append(resized_img)

    
    

```

```python

def lst_combined():
    all_maps = []
    lsts = list(codz_dic.values())
    for i in range(len(lsts)):
        for item in lsts[i]:
            if item in all_maps:
                remastered = item + " (BO3)"
                all_maps.append(remastered)
            all_maps.append(item)
    return all_maps

global maps_dic


global all_maps
all_maps = lst_combined()



all_maps.sort()
all_maps

maps_dic = {}




```

```python
#WARNING: Only run this cell once, unless you want abunch of images saved to your pc
corrected = []
for i in range(len(new_imgs)):
    imgs_rgb = cv2.cvtColor(new_imgs[i], cv2.COLOR_BGR2RGB)
    img_yes = Image.fromarray(imgs_rgb)
    img_yes.save(f"{all_maps[i]}.png")
    corrected.append(img_yes)
    
```

```python
codz_dic
def map_roulette():
    lsts = list(codz_dic.values())
    choice = input("Welcome to Map Roullette! Choose all to randomly choose from all games or enter a game to choose from a specific game!   ")
    if choice.upper() == 'ALL':
        maps = []
        for lst in lsts:
            for map in lst:
                if map not in maps:
                    maps.append(map)
        return random.choice(maps)
    elif choice.upper() in ['BO1', 'BLACK OPS', 'BLACK OPS 1']:
        return random.choice(lsts[0])
    elif choice.upper() in ['BO2','BLACK OPS 2', 'BLOPS2']:
        return random.choice(lsts[1])
    elif choice.upper() in ['BO3', 'BLACK OPS 3', 'BLOPS3']:
        return random.choice(lsts[2])
    elif choice.upper() in ['BO4', 'BLACK OPS 4', 'BlOPS 4']:
        return random.choice(lsts[3])
    elif choice.upper() in ['BOCW', 'BLACK OPS COLD WAR', 'COLD WAR', 'CW']:
        return random.choice(lsts[4])
    else:
        raise KeyError('GAME NOT FOUND! Please try again.   ')
    
    
```
